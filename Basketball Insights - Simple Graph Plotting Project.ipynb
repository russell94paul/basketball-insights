```python
def myplot(playerList):
    for name in playerList:
        plt.plot(Games[Pdict[name]], c = 'Black', ls='--', marker = 's', ms = 7, label = name)
    
    plt.legend(loc = 'upper right', bbox_to_anchor = (1.3,1))
    plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.show()
    
```

```python
def myplot(playerList):
    Col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Purple","KevinDurant":"Green","DerrickRose":"Brown","DwayneWade":"Yellow"}
    Marker = {"KobeBryant":"o","JoeJohnson":"s","LeBronJames":"p","CarmeloAnthony":"^","DwightHoward":"D","ChrisBosh":"o","ChrisPaul":"s","KevinDurant":"p","DerrickRose":"^","DwayneWade":"D"}
    for name in playerList:
        plt.plot(Games[Pdict[name]], c = Col[name], ls='--', marker = Marker[name], ms = 7, label = name)
    
    plt.legend(loc = 'upper right', bbox_to_anchor = (1.3,1))
    plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.show()
```
