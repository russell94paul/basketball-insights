```python
# Creating a plotting function
def myplot (data, playerlist = Players):
    
    Col = {"KobeBryant": "Black", "JoeJohnson": "Red", "LeBronJames": "Green", "CarmeloAnthony": "Blue", "DwightHoward": "Magenta", "ChrisBosh": "Yellow", "ChrisPaul": "Blue", "KevinDurant": "Pink", "DerrickRose": "Teal", "DwayneWade": "Grey"}
    Mrk = {"KobeBryant": "s", "JoeJohnson": "o", "LeBronJames": "^", "CarmeloAnthony": "D", "DwightHoward": "s", "ChrisBosh": "o", "ChrisPaul": "^", "KevinDurant": "D", "DerrickRose": "s", "DwayneWade": "o"}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], c = Col[name], ls = '--', marker = Mrk[name], ms = '7', label = name)
    # Moving this code below into the loop generates each plot for each player on a separate graph    
    plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.show()
```

```python
# Visualizing the FreeThrows Matrix
myplot(FreeThrows)
```

```python
# Visualizing the FreeThrowsAttempts Matrix
myplot(FreeThrowsAttempts)
```
